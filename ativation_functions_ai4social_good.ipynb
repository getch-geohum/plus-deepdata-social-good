```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def sigmoid(x):
    s=1/(1+np.exp(-x))
    ds=s*(1-s)  
    return s,ds
x=np.arange(-6,6,0.01)
sigmoid(x)
fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x,sigmoid(x)[0], color="#307EC7", linewidth=3, label="sigmoid")
ax.plot(x,sigmoid(x)[1], color="#9621E2", linewidth=3, label="derivative")
ax.legend(loc="upper right", frameon=False)
plt.show()
# fig.show()
```

```python
def tanh(x):
    t=(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x))
    dt=1-t**2
    return t,dt
z=np.arange(-4,4,0.01)
tanh(z)[0].size,tanh(z)[1].size
fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['bottom'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(z,tanh(z)[0], color="#307EC7", linewidth=3, label="tanh")
ax.plot(z,tanh(z)[1], color="#9621E2", linewidth=3, label="derivative")
ax.legend(loc="upper right", frameon=False)
plt.show()
```

```python
def ReLU(x):
    val = np.zeros(x.shape, x.dtype)
    rl=np.fmax(val, x)
    ds=rl*(1-rl)  
    return rl,ds
x=np.arange(-6,6,0.01)
fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x,ReLU(x)[0], color="#307EC7", linewidth=3, label="ReLu")
ax.plot(x,ReLU(x)[1], color="#9621E2", linewidth=3, label="derivative")
ax.legend(loc="upper right", frameon=False)
plt.show()
# fig.show()
```

```python
def LReLU(x):
    val = 0.01*x
    rl=np.fmax(val, x)
    ds=rl*(1-rl)  
    return rl,ds
x=np.arange(-6,6,0.01)
fig, ax = plt.subplots(figsize=(9, 5))
ax.spines['left'].set_position('center')
ax.spines['right'].set_color('none')
ax.spines['top'].set_color('none')
ax.xaxis.set_ticks_position('bottom')
ax.yaxis.set_ticks_position('left')
ax.plot(x,LReLU(x)[0], color="#307EC7", linewidth=3, label="sigmoid")
ax.plot(x,LReLU(x)[1], color="#9621E2", linewidth=3, label="derivative")
ax.legend(loc="upper right", frameon=False)
plt.show()
# fig.show()
```

```python
# softmax
def softMax(x):
    if len(x.shape) == 1:
        return np.exp(x)/np.sum(np.exp(x))
    else:
        return np.exp(x)/np.sum(np.exp(x), axis=1).reshape(-1,1)
    
mm = np.array([[1.8, 0.9, 0.68], [1.8, 0.9, 2.68], [1.8, 3.9, 0.68]])
my = softMax(mm)
print(my)
```
